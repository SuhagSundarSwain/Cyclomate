```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Creating blank array i.e black image

```python
blank_img = np.zeros( (600,600) )
```

```python
plt.imshow(blank_img, cmap= 'gray')
```

### Adding Text to Blank Image

```python
text_img = cv2.putText(
            blank_img,
            text='ABCDE',
            org = (50,300),
            fontFace=cv2.FONT_HERSHEY_SIMPLEX,
            fontScale=5,
            color=(255,255,255),
            thickness=25
           )
```

```python
plt.imshow(text_img, cmap='gray')
```

```python
kernel = np.ones( (5,5), dtype=np.uint8 )
```

```python
kernel
```

### EROSION

#### Take kernel matrix and compare colors and clean noises

```python
erode_img = cv2.erode(text_img, kernel)
```

```python
plt.imshow(erode_img, cmap='gray')
```

```python
erode_img = cv2.erode(text_img, kernel, iterations = 3)
```

```python
plt.imshow(erode_img, cmap='gray')
```

```python
erode_img = cv2.erode(text_img, kernel, iterations = 4)
```

```python
plt.imshow(erode_img, cmap='gray')
```

### DILUSION

```python
dilate_img = cv2.dilate(erode_img, kernel, iterations = 5)
```

```python
plt.imshow(dilate_img, cmap='gray')
```

```python
dilate_img = cv2.dilate(erode_img, kernel, iterations = 4)
```

```python
plt.imshow(dilate_img, cmap='gray')
```

```python
dilate_img = cv2.dilate(erode_img, kernel, iterations = 3)
```

```python
plt.imshow(dilate_img, cmap='gray')
```

## MORPHOLOGY

### morphologyEx

#### open

```python
morph_open_img = cv2.morphologyEx(src = text_img, kernel = kernel,op = cv2.MORPH_OPEN)
```

```python
plt.imshow(morph_open_img, cmap='gray')
```

### Nothing Much effect here : deepend on image to image

#### close

```python
morph_close_img = cv2.morphologyEx(src = text_img, kernel = kernel,op = cv2.MORPH_CLOSE)
```

```python
plt.imshow(morph_close_img, cmap='gray')
```

#### Gradient

```python
morph_gradient_img = cv2.morphologyEx(text_img, cv2.MORPH_GRADIENT, kernel)
```

```python
plt.imshow(morph_gradient_img, cmap='gray')
```
